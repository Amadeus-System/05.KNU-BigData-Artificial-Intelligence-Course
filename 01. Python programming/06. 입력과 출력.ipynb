## 표준 입출력(Standard Input/Output)

### 표준 입력 함수
* `input()`함수:콘솔 창을 통해서 사용자 입력을 받는 표준 입력 함수
* `input()`함수 안에 입력 받기 위한 질문을 텍스트로 넣을 수 있음


```
name = input("이름: ")
print(name)
```

#### [Lab] 섭씨 온도를 화씨 온도로 변환

* 섭씨 온도를 화씨 온도로 변환하는 공식: $F = (C * 1.8) + 32$

```
celsius = float(input("섭씨 온도: "))
fahrenheit = (celsius * 1.8) + 32
print("화씨 온도:", fahrenheit)
```

#### [Lab] 구구단 중의 하나의 단을 입력받아 계산

* 출력할 단을 입력
* 해당 단의 계산 결과 출력

```
i = int(input("출력할 단: "))
for j in range(1, 10):
    print("{0} x {1} = {2}".format(i, j, i * j))
```

### 표준 출력 함수

* `print()`함수:콘솔 창을 통해서 결과를 출력하는 표준 출력 함수
* sep='separator':개체가 분리하는 방법 지정, 기본값은 ''
* end='end':끝에 출력할 항목 지정, 기본값은 '\n'
* file:쓰기 방법이 있는 객체, 기본값은 sys.stdout
* flush:True일 경우 flush하고, False일 경우 버퍼, 기본값은 False

```
print("Hello Python")
print("안녕 파이썬")
print(7)
print(3.14)
print([1, 2, 3])

print("Hello", "Python", sep="---") # 컴마 구분을 ---로
print(1, 2, 3)
print(1, end=' ') # 줄바꿈 안하고 스페이스 처리함
print(2, end=' ')
print(3, end=' ')
```

#### [Lab] 구구단 출력

* `print()` 함수를 사용하여 각 단마다 횡으로 출력


```
for i in range(2, 10):
    for j in range(1, 10):
        print("{0}x{1}={2}".format(i, j, i * j), end='\t') # 태그
    print()
```

### 파일 입출력 과정

* 파일의 입출력 과정은 단계를 가짐
* 파일을 열고, 파일을 읽거나 쓰고, 파일을 닫는 순서(열었으면 항상 닫아야 함)

### 파일 열기/닫기

* file.txt 파일을 생성하고, 열고 닫기


```
f = open("file.txt", 'w') # 쓰기모드로 열기
f = open("file.txt", 'r') # 읽기모드로 열기, default
f.close()
```

```
!ls file.txt # 파일 생성 확인
```

### 파일 모드(File Mode)

|파일모드|설명|
|---|---|
|(생략)|r과 동일한 모드|
|r|읽기 모드(기본값)|
|w|쓰기 모드, 기존에 파일이 있으면 덮어쓰기|
|a|쓰기 모드, 기존에 파일이 있으면 이어서 쓰기(append)|
|+|읽기/쓰기 모드|
|t|텍스트 모드(기본값), 텍스트 파일을 처리|
|b|이진 모드, 이진 파일을 처리|

### 텍스트 파일 쓰기


#### write()

* `write()`를 이용하여 파일에 쓰기

```
f = open("file.txt", 'w')
f.write("Hello Python")
f.close()
```

```
!cat file.txt # 내용 확인
```

#### writelines()

* `writelines()`를 이용하여 list를 파일에 쓰기

```
li = ["One\n", "Two\n", "Three\n"]
f = open("file.txt", 'w')
f.writelines(li)
f.close()
```

```
!cat file.txt  # 내용 확인
```

#### 표준 입력 → 파일 쓰기

* 표준 입력을 `input()` 함수를 통해 입력 받고, `write()` 함수를 통해 파일에 쓰기

```
text = input("입력: ")
f = open("file.txt", 'w')
f.write(text)
f.close()
```

```
!cat file.txt
```

```
text = [str(text + '\n') for text in input("여러 값 입력: ").split()]
f = open("file.txt", 'w')
f.writelines(text)
f.close()
```

```
!cat file.txt
```

```
x = ""
text = ""
while x != 'q':
    x = input("반복 입력: ")
    text += x + '\n'

f = open("file.txt", 'w')
f.writelines(text)
f.close()
```

```
!cat file.txt
```

#### readline()

* `readline()`를 이용하여 텍스트 파일을 라인 단위로 읽고 화면에 출력

```
f = open("anthem.txt", 'r')
while True:
    line = f.readline()
    if not line:
        break  # 읽어왔는데 아무것도 없으면 중단
    print(line)

f.close()
```

#### readlines()

* `readlines()`를 이용하여 텍스트 파일의 여러 라인을 읽고 화면에 출력

```
f = open("anthem.txt", 'r')
lines = f.readlines() # 여러 줄 한번에 가져옴
print(lines)
f.close()
```

#### read()

* `read()`를 이용하여 텍스트 파일을 읽고 화면에 출력

```
f = open("anthem.txt", 'r')
data = f.read() 
print(data)
f.close()
```

#### tell()

* 파일 포인터 위치 변경

```
f = open("anthem.txt", 'r')
while True:
    print(f.tell()) # 파일 위치를 출력
    line = f.readline()
    if not line:
        break
    print(line)

f.close()
```

#### seek()

* 파일 포인터 위치 이동

```
f = open("anthem.txt", 'r')
f.seek(46) # 46번째 위치로 이동하여 read함
line = f.readline()
print(line)
f.seek(0)  # 다시 0번째로 되돌아와 read함
line = f.readline()
print(line)
f.close()  # 닫기 필수!
```

### with 문

* 항상 파일을 `open()` 함수로 열고, `close()` 함수로 닫아야 하는 일을 자동으로 처리
* with문을 이용하면 with 블록 내에서 파일을 열고 벗어나면 파일을 닫음

```
# 원래 위에서 하던 방식
f = open("file.txt", 'w')
f.write("Hello Python")
f.close()
```

```
# with문 이용
with open("file.txt", 'w') as f:
    f.write("Hello Python")
```

### [Lab] 표준 입력으로 받은 내용을 파일로 쓰기

* 사용자로부터 표준 입력 받음
* 받은 내용을 계속 파일 쓰기

### 이진 파일 복사

* 하나의 이진 파일을 읽어서 다른 이진 파일로 쓰기 (복사)

```
input = open("pythonfile.jpg", "rb") # 바이너리 파일로 읽는다.
output = open("pythonfile_copy.jpg", "wb") # 복사
while True:
    fp = input.read(1)  # file point에 하나씩 읽어온다.
    if not fp:
        break  # 읽어올 게 없으면 중단
    output.write(fp) # 읽어온 자료 기록

input.close()
output.close()
```

### 디렉토리 및 파일 확인

* `isdir()`: 디렉토리 존재 확인
* `isfile()`: 파일 존재 확인
* `exists()`: 디렉토리/파일 존재 확인

```
import os.path
print(os.path.isdir('test'))  # test 디렉터리 있는가 출력
print(os.path.isdir('test/1')) 
print(os.path.isdir('test/4')) 
print(os.path.isfile("pythonfile.jpg")) # 파일 확인
print(os.path.isfile("pythonfile_copy.jpg")) 
print(os.path.isfile("pythonfile_copy2.jpg")) 
print(os.path.exists('test2')) 
print(os.path.exists('test2/1')) 
print(os.path.exists('test2/4')) 
# dir은 디렉터리에 대해서만 확인
# exist는 디렉터리와 파일까지
```

### 디렉토리 목록 보기

* `os.walk()`: 디렉토리 목록 보기

```
import os
for dir_name, dir_list, file_names in os.walk('./'):  # 현재 디렉터리에서만
    for file_name in file_names:
        print(os.path.join(dir_name, file_name)) # join : 결합
```

### 파일 압축

* zipfile : 압축 관련 라이브러리
* `ZipFile()` : 압축 파일 열기
* `write()` : 압축 파일에 쓰기
* `extractall()` : 전체 압축 해제
* `close()` : 압축 파일 닫기



```
import zipfile
comp = zipfile.ZipFile("new.zip", 'w') # 쓰기모드로 압축파일 열기
comp.write("pythonfile.jpg")
comp.close()
```

```
!ls
```

```
decomp = zipfile.ZipFile('new.zip', 'r')
decomp.extractall('new')  # 압축풀 디렉터리명 new
decomp.close()
```

```
!ls new # 확인
```
