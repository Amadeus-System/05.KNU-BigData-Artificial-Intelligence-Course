# Google Colaboratory

### Google Colab 특징
* 구글 클라우드 기반의 무료 개발 환경 서비스
* 환경설정 및 실행까지 매우 빠른 환경
* 딥러닝 실행이 가능한 정도의 고사양 환경 제공
* Jupyter Notebook 환경 제공
* 대부분의 패키지들이 이미 설치된 환경 제공
* 여러 사용자와 동시에 사용 가능
* PC, 태블릿, 모바일 상관없이 인터넷 브라우저만 있으면 언제 어디서나 접속 가능
* 목차나 Markdown 미리보기 등 다양한 기능 제공
* 구글 드라이브와 연동이 가능
* Git이나 Github와 쉽게 연동 가능

##와일드카드(wildcard) 매칭

* 단어의 중간이나 마지막 글자로 매칭하여 객체나 속성 찾기

## 매직 명령어(magic commands)

* Jupyter Notebook 환경에서 파이썬 파일 저장 및 실행

````
%%writefile test.py
print("Hello Colab")
````

````
%run test.py # 실행
````

* 코드 실행 시간 측정

````
%%timeit?
````

````
%timeit li = [n ** n for n in range(10000)]  # 한 줄 코드일 경우
````

````
%%timeit  
# 여러 줄 코드일 경우
li = []
for n in range(10000):
    li.append(n ** n)
````

## 입력과 출력 이력

* In과 Out 객체를 통해 이전 명령어와 명령어의 출력 결과 보기

````
In
# 그 동안의 모든 명령기록이 히스토리로 저장되어 있음
````

````
In[26]
# 26번째 실행
````

````
Out
# In에 대한 결과물 저장 히스토리
````

````
Out[41]
````

````
print("In[41]: " + In[41] + "\nOut[41]: " + Out[41])
````

* `_`를 이용해 이전 출력값 출력하기

````
print(_) # 이전 출력값
````

````
print(__) # 이전이전 출력값
````

````
print(___) # 이전이전이전 출력값
````

````
_41
# 41번째 출력하기
````

* `%history`를 이용한 입력 이력 살펴보기

````
%history -n 1-7  # 1번부터 7번까지 명령수행한 것 히스토리 보기
````

* `%rerun`을 이용해 이전 입력 이력 다시 실행

````
%rerun
# 이전 입력 다시 실행
````

## 셸 명령어

* 텍스트 기반의 셸 명령어 처리
* `!` 문자를 명령어 앞에 붙여서 셸 명령어 사용 가능

* `ls`: 디렉토리 리스트 보기

````
!ls
````

* `pwd`: 현재 경로 보기

````
!pwd
````

* `cd`: 디렉토리 변경
* IPython에서는 임시 셸에서 실행

````
!cd sample_data/ && ls
# 디렉터리 바꾸자마자 ls 실행
````

* `%cd`: 지속적인 디렉토리 변경

````
%cd sample_data/
````

* `echo`: 화면 출력

````
!echo "Shell"
````

* `mkdir`: 디렉토리 생성

````
!mkdir tmp  # tmp 파일 생성
````

````
!ls # 만들어졌나 확인
````

* `cat`: 파일 보기

````
!cat README.md
````

* `cp`: 디렉토리/파일 복사

````
!cp README.md tmp/  # README를 tmp 디렉터리에 복사
````

````
!ls tmp/
````

* `rm`: 디렉토리/파일 삭제

````
!rm -r tmp # tmp 폴더와 그 안에 파일 다 지우기
````
