## 리스트(List)

* 변경 가능한 시퀀스 자료형
* 하나의 변수에 여러 값 할당 가능
* 같은 자료형이 아니라 다른 자료형을 가지는 값들을 포함 가능
* 일반적으로 비슷한 항목들의 모음을 순서대로 저장하는데 사용




```
print([])
print([1, 2, 3])
print(['One', 'Two', 'Three'])
print([1, 'One', 2, 'Two', 3, 'Three'])
print([1, 2, 3, ['One', 'Two', 'Three']])
```

### 리스트 인덱싱(List Indexing)

* 리스트의 각 위치에 해당하는 값에 접근하기 위해서 주소 개념의 숫자를 사용

```
li = ['One', 'Two', 'Three']
print(li)
print(li[0])
print(li[1])
print(li[2])
print(li[-1])
print(li[-2])
print(li[-3])
```

### 중첩 리스트 인덱싱(Nested List Indexing)

* 리스트 안에 리스트가 있을 경우 인덱스 접근 방법

```
li = [1, 2, 3, ['One', 'Two', 'Three']]
print(li)
print(li[3])
print(li[3][0])
print(li[3][1])
print(li[3][2])
```

### 리스트 슬라이싱(List Slicing)

* 리스트의 인덱스를 통해서 부분만 가져올 때 사용

```
li = ['One', 'Two', 'Three']
print(li)
print(li[1:2])
print(li[1:])
```

### 중첩 리스트 슬라이싱(Nested List Slicing)

* 중첩된 리스트에서 일부분만 인덱스를 통해서 가져올 때 사용

```
li = [1, 2, 3, ['One', 'Two', 'Three']]
print(li)
print(li[2:4])
print(li[3][2:])
print(li[3][:2])
```

### 리스트 연산자/함수(List Operators/Function)

* 더하기 + 연산자
* 곱하기 * 연산자
* 리스트 길이를 구하는 `len()` 함수

```
list_1 = ['One', 'Two', 'Three']
list_2 = ['Four', 'Five', 'Six']

print(list_1)
print(list_2)
print(list_1 + list_2)  # concatenate 연산
print(list_1 * 3)  # 반복 연산
print(len(list_1))
print(len(list_1 * 3))
```

### 리스트 수정(List Modify)

* 리스트 인덱스 주소를 이용한 값 수정

```
li = ['One', 'Two', 'Three']
print(li)
li[2] = 3
print(li)
li[1] = 2
print(li)
li[0] = 1
print(li)
```

### 리스트 메소드(List Methods)


#### append()

* 리스트 요소 추가

```
li = ['One', 'Two', 'Three']
li.append('Four')
print(li)
li.append([1, 2, 3, 4])
print(li)
```

#### sort()

* 리스트 정렬

```
li = [10, 40, 20, 30]
print(li)
li.sort()
print(li)
li = ['orange', 'apple', 'banana', 'strawberry']
print(li)
li.sort()
print(li)  # 알파벳 순 정렬
```

#### reverse()

* 리스트 요소 반전

```
li = ['orange', 'apple', 'banana', 'strawberry']
print(li)
li.reverse()
print(li)
```

#### index()

* 리스트의 요소 값에 대한 인덱스를 반환

```
li = [10, 40, 20, 30]
print(li)
print(li.index(10))
print(li.index(20))
```

#### insert()

* 리스트 요소 삽입

```
li = [10, 40, 20, 30]
print(li)
li.insert(4, 50)
print(li)
li.insert(0, 60)
print(li)
```

#### remove()

* 리스트 요소 제거

```
li = [10, 40, 20, 30]
print(li)
li.remove(40)
print(li)
li.remove(30)
print(li)
```

#### del

* 리스트 요소 제거 연산

```
li = [10, 40, 20, 30]
print(li)
del li[0]
print(li)
del li[2]
print(li)  # 삭제할수록 인덱스 값이 바뀜에 주의!
```

#### pop()

* 리스트 요소를 방출

```
li = [10, 40, 20, 30]
print(li)
li.pop()
print(li)
li.pop(0)
print(li)
```

#### count()

* 리스트 요소의 갯수 계산

```
li = [10, 20, 20, 30, 30]
print(li)
print(li.count(30))
print(li.count(20))
```

#### extend()

* 리스트 확장

```
li = [10, 40, 20, 30]
print(li)
li.extend([50, 60])
print(li)
```

## 튜플(Tuple)

* 리스트와 유사하지만 변경 불가능한 시퀀스 자료형
* 하나의 변수에 여러 값 할당 가능
* ‘(‘와 ‘)’를 사용하여 표현

```
print(())
print((1, 2, 3))
print(('One', 'Two', 'Three'))
print((1, 'One', 2, 'Two', 3, 'Three'))
print((1, 2, 3, ('one', 'two', 'three')))
```

### 튜플 인덱싱(Tuple Indexing)

* 튜플의 각 위치에 해당하는 값에 접근하기 위해서 주소 개념의 숫자를 사용

```
tu = ('one', 'two', 'three')
print(tu)
print(tu[0])
print(tu[-1])
```

### 중첩 튜플 인덱싱(Nested Tuple Indexing)

* 튜플 안에 튜플이 중첩되어 있을 경우, 인덱스 접근 방법

```
tu = (1, 2, 3, ('one', 'two', 'three'))
print(tu)
print(tu[3])
print(tu[3][0])
print(tu[3][1])
print(tu[3][2])
```

### 튜플 슬라이싱(Tuple Slicing)

* 튜플의 인덱스를 통해서 부분만 가져올 때 사용

```
tu = ('one', 'two', 'three')
print(tu)
print(tu[0:])
print(tu[1:2])
print(tu[1:])
```

### 중첩 튜플 슬라이싱(Nested Tuple Slicing)

* 중첩된 튜플에서 일부분만 인덱스를 통해서 가져올 때 사용

```
tu = (1, 2, 3, ('one', 'two', 'three'))
print(tu)
print(tu[3])
print(tu[3][2:])
print(tu[3][:2])
```

### 튜플 연산자/함수(Tuple Operators/Function)

* 더하기 + 연산
* 곱하기 * 연산
* 튜플 길이를 구하는 `len()` 함수



```
tuple_1 = ('one', 'two', 'three')
tuple_2 = ('four', 'five', 'six')
print(tuple_1)
print(tuple_2)
print(tuple_1 + tuple_2)
print(tuple_1 * 3)
print(len(tuple_1))
print(len(tuple_1 + tuple_2))
```

## 세트(Set)

* 데이터 중복을 허용하지 않는 구조
* 순서가 없는 데이터 집합을 위한 구조
* 인덱싱으로 값을 접근할 수 없음

```
# 집합은 중복,순서 없다. 순서가 없으니 인덱싱도 안먹힘! (중요)
print({})
print({"One", "Two", "Three"})
print({10, 20, 30, 40})
```

### 세트 연산자(Set Operators)

* 교집합: &
* 합집합: |
* 차집합: -
* 여집합: ^

```
set_1 = {10, 20, 30, 30}
set_2 = {30, 30, 40, 50}
print(set_1)
print(set_2)  # 집합은 중복허용 안됨
print(set_1 & set_2) # 교집합
print(set_1 | set_2) # 합집합
print(set_1 - set_2) # 차집합
print(set_1 ^ set_2) # 여집합
```

### 세트 메소드(Set Methods)

* 교집합: intersection()
* 합집합: union()
* 차집합: difference()
* 여집합: symmetric_difference()

```
set_1 = {10, 20, 20, 30}
set_2 = {30, 30, 40, 50}
print(set_1)
print(set_2)
print(set_1.intersection(set_2))
print(set_1.union(set_2))
print(set_1.difference(set_2))
print(set_1.symmetric_difference(set_2))
```

* 요소 추가: add()
* 여러 요소 추가: update()
* 요소 제거: remove()
* 요소 제거: discard()
* 모든 요소 제거: clear()

```
set_1 = {10, 20, 30, 40}
print(set_1)

set_1.add(50)
print(set_1)

set_1.update([60, 70])
print(set_1)

set_1.remove(70)
set_1.remove(60)
print(set_1)

set_1.discard(30)
print(set_1)

set_1.clear()
print(set_1)
```

## 딕셔너리(Dictionary)

* 키(key)와 값(value)의 쌍으로 구성된 데이터
* 순서가 없는 데이터
* 키를 통해 값을 얻음
* 동일한 키가 있을경우 덮어씀

```
dic = {1:'One', 2:'Two', 3:'Three'}
print(dic)

dic = {1:'One', 2:'Two', 3:'Three', 1:'One', 2:'Two', 3:'Three'}
print(dic)
```

### 딕셔너리 요소 추가/삭제

* 딕셔너리의 해당 키 값에 값을 추가하여 요소 추가
* del을 이용하여 요소 제거

```
dic = {1:'One', 2:'Two', 3:'Three'}
print(dic)
print(dic[2]) # 인덱스가 아니라 키 접근임, 헷갈리지 말 것!
dic[4] = 'Four'  # 키와 값 추가
print(dic)
dic[5] = 'Five'
print(dic)
del dic[4]  # 키4와 그에 해당하는 값 제거
print(dic)
```

### 딕셔너리 메소드(Dictionary Methods)


```
dic = {1:'One', 2:'Two', 3:'Three'}
print(dic)
print(dic.keys())
print(dic.values())
print(dic.items())
print(dic.get(2)) # 2라는 키에 해당하는 값 얻기
print(dic.pop(3)) # 키에 해당하는 값 추출(빼냄)
print(dic) # 빠져있음 확인
dic.clear()
print(dic)
```
