## 입력값이 양수, 음수, 0인지 구분하는 코드를 작성하시오. (20점)

```
n = int(input("정수 입력:"))

# 코드 작성
if n > 0 :
    print("양수입니다.")
elif n < 0:
    print("음수입니다.")
elif n == 0:
    print("0입니다.")
```

## 주민등록번호를 입력받아 남성과 여성을 구분하는 코드를 작성하시오. (20점)

```
reg_num = input("주민등록번호(숫자만 입력하세요.):")

#코드 작성
if reg_num[6] == '1':  # 문자열은 시퀀스이고, 문자열의 성분도 문자!
    print("남자입니다.")
elif reg_num[6] == '2':
    print("여자입니다.")
```

## 삼각형과 사각형의 넓이는 구하는 함수를 작성하시오. (20점)

```
def triangle(w, h):
  return 0.5 * w * h

def rectangle(w, h):
  return w * h

width = int(input('너비 입력:'))
height = int(input('높이 입력:'))

#코드 작성
#삼각형과 사각형 넓이 출력
print("삼각형의 넓이:", triangle(width, height))
print("사각형의 넓이:", rectangle(width, height))
```

## 두 수를 입력받아 계산하는 함수들을 작성하시오. (20점)

```
def add(a, b) :
	return a + b

def sub(a, b) :
	return a - b

def mult(a, b) :
	return a * b

def div(a, b) :
	return a / b

x = int(input('정수 x: '))
y = int(input('정수 y: '))

print('두 수의 덧셈 : %d' % add(x, y))
print('두 수의 뺄셈 : %d' % sub(x, y))
print('두 수의 곱셈 : %d' % mult(x, y))
print('두 수의 나눗셈 : %d' % div(x, y))
```

## 타자 게임 코드를 발전시켜 보세요. (20점)

```
import random
import time

# 단어 리스트 w 선언과 초기화!!
w = ['cat', 'bird','dog', 'fox', 'monkey', 'lion', 'tiger', 'mouse', 'panda', 'frog', 'snake', 'wolf', 'horse']

n = 1 # 문제 번호로 사용될 변수

print('[타자 게임] 준비되면 엔터를 누르세요.')
input()

start = time.time() # 엔터 누르고 나서의 현재 시간을 start에 대입!

q = random.choice(w) # 리스트 w 내의 값 중 하나가 선택

while True :  # 무한루프로 문제 내기
    print('** 문제', n, '번 **')
    print(q)
    x = input()
    
    if x == 'quit':
        print("[타자 게임]을 종료합니다^^.")
        break # ======= 무한루프 종료 ========

    elif q == x :  # 문제와 답이 같으면 통과시킴
        print('통과')
        n = n + 1 # 문항 수 1 증가
        q = random.choice(w) # 리스트 w의 다른 내용으로 q에 대입
    else :
        print('오타~ 다시 도전!')
        

end = time.time()
et = end - start
et = format(et, '.2f') # et의 값을 소수점 2자리 실수로 표현!

print('타자 시간 :', et, '초')
```
