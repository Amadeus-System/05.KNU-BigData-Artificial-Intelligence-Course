## 1부터 100 사이의 홀수와 짝수의 누적합을 구하시오. (10점)

```
even_sum = 0
odd_sum = 0

for i in range(1, 101):
  # 코드 입력
  if i % 2 == 0:  # 2로 나눠서 나머지 0이면 짝수
      even_sum += i 
  elif i % 2 != 0: # 2로 나눠서 나머지 0이 아니면 홀수
      odd_sum += i

print("짝수의 누적합:", even_sum)
print("홀수의 누적합:", odd_sum)
```

## 2부터 100 사이의 소수를 구하는 코드를 작성하시오. (30점)

```
def judgement(N):
    # 입력 N은 2이상의 자연수로만 한정한다.
    for i in range(2, N):  # 1과 자기자신 N을 제외하고 나누어진다면 소수가 아니다.
        if N % i == 0:
            return False
    else:
        return True # 그 이외에는 소수이다.


prime_numbers = []
for i in range(2, 101):
    # 코드 입력
    if judgement(i) == True:  # 소수 판정 결과 True이면
        prime_numbers.append(i)  # 그 수를 prime_numbers 리스트에 넣는다.

print(prime_numbers)
  
```

## 나이에 따른 입장료를 계산하는 코드를 작성하시오. (30점)

* 1~7세: 1000원
* 8~20세: 2000원
* 21~40세: 3000원
* 41~세: 1500원

```
age = 30
pay = 0

# 코드 입력
def cal(age):  # 나이에 따라 입장료 계산하는 함수
    if 1 <= age <= 7:
        return 1000
    elif 8 <= age <= 20:
        return 2000
    elif 21 <= age <= 40:
        return 3000
    elif age >= 41:
        return 1500

pay = cal(age)
print("나이: {0}, 입장료: {1}".format(age, pay))
```
