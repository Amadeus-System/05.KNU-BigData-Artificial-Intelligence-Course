# 네이버 웹툰 스크래핑

## Selenium 및 웹 드라이버 설치

```
# 설치
!pip install Selenium
# 업데이트
!apt-get update
# 웹드라이버 설치
!apt install chromium-chromedriver
# /usr/bin에 카피
!cp /usr/lib/chromium-browser/chromedriver /usr/bin

import sys
sys.path.insert(0, '/usr/lib/chromium-browser/chromedriver' ) # 경로설정

from selenium import webdriver  # 웹 드라이버 import

# 클라우드(colab)상에서 동작하므로 웹 브라우저를 일반적으로 하듯이 띄울 수 없음
# 브라우저 띄우지 않고, 자동화를 한다는 뜻
# 그에 따라 여러 옵션 disable함
chrome_options = webdriver.ChromeOptions()  
chrome_options.add_argument('--headless')  
chrome_options.add_argument('--no-sandbox') 
chrome_options.add_argument('--disable-dev-shm-usage')
```

## 네이버 웹툰 페이지 로드

```
wd = webdriver.Chrome('chromedriver', options = chrome_options) # 웹 드라이버 객체 생성
# 옵션에 위에서 지정한 chrome_options 넣어줌

# url을 이용하여 get 메서드로 접속함
wd.get('https://comic.naver.com/index.nhn')
```

## 장르별 추천 웹툰 긁어오기

```
# class name이 title인 것을 모두 가져옴
for title in wd.find_elements_by_class_name("title"): 
    print(title.text)
```

```
import time

genre_ul = wd.find_element_by_class_name("tab_gr") # 단일 클래스 선택
genre_list = genre_ul.find_elements_by_tag_name('a') 

for genre in genre_list:

    genre.click() # 클릭 메소드
    time.sleep(1) # 클릭 이후 잠깐 시간지연
    print("[",genre.text,"]")


    genre_rec_list = wd.find_elements_by_class_name("genreRecomInfo2")
    
    for genre_rec in genre_rec_list:
        title_class = genre_rec.find_element_by_class_name("title")
        title = title_class.find_element_by_tag_name('a').text
        user = genre_rec.find_element_by_class_name("user").text
        print("\t", title, "-", user)
        # 프로그램이 너무 빠름, 클릭 이후 refresh time보다 더 빠르면 에러 발생
```
